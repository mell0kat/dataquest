```python
import pandas as pd
unrate = pd.read_csv('data/Hawaii_Unemployment.csv')
unrate.dtypes
unrate['Month/Year'] = pd.to_datetime(unrate['Month/Year'])

first_twelve = unrate[0:12]
first_twelve.dtypes
```

```python
import matplotlib.pyplot as plt 

plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('HW Monthly Unemployment Trends, 1990')
plt.xticks(rotation=90)
plt.plot(first_twelve['Month/Year'], first_twelve['Unemployment Rate'])
plt.show()
```
