```python
autos['brand'].unique()
```

```python
most_freq_brands = autos['brand'].value_counts(normalize=True)[autos['brand'].value_counts(normalize=True)>0.05].index
```

```python
mean_price_by_brand = {}
for brand in most_freq_brands:
    autos_for_brand = autos[autos['brand'] == brand]
    mean_price_by_brand[brand] = autos_for_brand['price'].mean()
mean_price_by_brand
```

### Aggregating price by brand for most frequest brands
* Audi, bmw, mercedes are expensive brands, each with average used auto price > $8000
* VW is mid-range, whereas Ford and Opel are more affordable

```python
mean_mileage_by_brand = {}
for brand in most_freq_brands:
    autos_for_brand = autos[autos['brand'] == brand]
    mean_mileage_by_brand[brand] = autos_for_brand['odometer_km'].mean()
mean_mileage_by_brand
```

```python
mean_price_by_brand_series = pd.Series(mean_price_by_brand)
mean_price_by_brand_series
```

```python
mean_mileage_by_brand_series = pd.Series(mean_mileage_by_brand)
mean_mileage_by_brand_series
```

```python
mean_price_by_brand_df = pd.DataFrame(mean_price_by_brand_series, columns=['mean_price'])
```

```python
mean_price_by_brand_df['mean_mileage'] = mean_mileage_by_brand_series
mean_price_by_brand_df
```

### Aggregating price and mileage by brand
    * The top brands have approximately the same mileage
    * Mileage oes not appear to have any significant effect on price
